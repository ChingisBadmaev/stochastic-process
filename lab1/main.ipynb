```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# вычисление интеграла методом трапеций
def trapezoid_rule(func, a, b, nseg):
    """Правило трапеций
       nseg - число отрезков, на которые разбивается [a;b]"""
    dx = 1.0 * (b - a) / nseg
    sum = 0.5 * (func(a) + func(b))
    for i in range(1, nseg):
        sum += func(a + i * dx)

    return sum * dx
```

```python
# детерминированная часть
def part_det(t):
    return -0.1 * t - ((math.exp(t) / 100) ** 2) / 2
```

```python
def second_integrate(a, b, nseg):
    sum = 0
    khi = np.random.normal(0, 1, 1)[0]
    for t in range(0, nseg):
        x_t = np.random.uniform(a + t / nseg, a + (t + 1)  / nseg)
        sum = math.exp(x_t)
    return sum * math.sqrt(a) * khi / nseg
```

```python
ans = []
for i in range(0, 8):
    nseg_1 = 2500
    nseg_2 = 25000000
    part_determ = trapezoid_rule(part_det, i * (1 / 2), (i + 1) * (1 / 2), nseg_1)
    part_stoch = second_integrate(1/2, 1, nseg_2)
    S_0 = 35
    S_t = S_0 * math.exp(part_determ + part_stoch)  
    ans.append(S_t)
ans.insert(0, 35)
print(ans)
```

```python
x = [0, 1/2, 1, 1 + 1/2, 2, 2 + 1/2, 3, 3 + 1/2, 4]
y1 = ans

plt.figure(figsize=(12, 7))

plt.plot(x, y1, 'o-r', alpha=0.7, label="risk asset", lw=5, mec='b', mew=2, ms=10)
plt.legend()
plt.grid(True)
```

### Цены облигаций

```python
# детерминированная часть
def part_det(t):
    return -0.1 * t


ans_2 = []
nseg = 2500
for i in range(0, 8):
    part_determ = trapezoid_rule(part_det, i * (1 / 2), (i + 1) * (1 / 2), nseg)
    S_0 = 35
    S_t = S_0 * math.exp(part_determ)
    ans_2.append(S_t)
ans_2.insert(0, 35)
print(ans_2)
```

```python
x = [0, 1/2, 1, 1 + 1/2, 2, 2 + 1/2, 3, 3 + 1/2, 4]
y1 = ans_2

plt.figure(figsize=(12, 7))
plt.plot(x, y1, 'v-.g', label="bond", mec='r', lw=2, mew=2, ms=12)
plt.legend()
plt.grid(True)
```
