```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def runge_kutta(y, x, dx, f):
    k1 = f(y, t)
    k2 = f(y + 0.5 * k1, x + 0.5 * dx)
    
    k3 = f(y + 0.5 * k2, x + 0.5 * dx)
    k4 = f(y + k3, x + dx)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6.

def func(y, t):
    nseg = 1000
    x_t = np.random.uniform(0 + t / nseg, 0 + (t + 1)  / nseg)
    return y + -0.1 * t * y * (1 / 250) + (math.exp(t) / 100) * y * math.sqrt(t) * x_t

t = 0.0
y = 35.0
dt = 4 / 1000
ys, ts = [], []

while t <= 10:
    y = runge_kutta(y, t, dt, func)
    t += dt
    ys.append(y)
    ts.append(t)

plt.plot(ts, ys, label='runge_kutta')

plt.legend()
plt.show()
```
